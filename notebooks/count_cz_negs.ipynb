```python
import polars as pl
import numpy as np
from src.utils import construct_sparse_adj_mat
from pathlib import Path
from itertools import product
```

```python
adj, _, _ = construct_sparse_adj_mat(
    Path("/home/stef/quest_data/hiec/data/sprhea/v3_folded_pt_ns.csv")
)
```

```python
nz_i, nz_j = adj.nonzero()
nnz = len(nz_i)
adj = np.array(adj.todense())
```

```python
batch_size = 64
epochs = 30
rng = np.random.default_rng(seed=1234)
```

```python
cum_negatives = set()
for ep in range(epochs):
    batch_idxs = rng.choice(nnz, size=(nnz // batch_size, batch_size), replace=False)
    for batch in batch_idxs:
        i = nz_i[batch]
        j = nz_j[batch]
        cart_i, cart_j = tuple(zip(*product(i, j)))
        negatives = (adj[cart_i, cart_j] == 0)
        neg_i = np.array(cart_i)[negatives]
        neg_j = np.array(cart_j)[negatives]
        this_negs = set(zip(neg_i, neg_j))
        cum_negatives.update(this_negs)

    print(f"{len(cum_negatives):,} unique negative samples by epoch {ep+1}")

    if len(cum_negatives) > 100_000_000:
        break
```
